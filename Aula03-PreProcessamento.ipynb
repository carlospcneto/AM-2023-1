```python
import numpy as np
import pandas as pd

# Localização do arquivo
filepath = 'data/Iris_Data.csv'
# Importando os dados
data = pd.read_csv(filepath)
# Amostragem de 5 colunas sem reposição
sample = (data.sample(n=5,replace=False,random_state=42))
print(sample.iloc[:,-3:])
```

```python
from sklearn.decomposition import PCA

#Cria uma instância da classe PCA
PCAinst = PCA(n_components=2, whiten=True)

#Ajusta a instância aos dados e o transforma
trans = PCAinst.fit_transform(data.iloc[0:,:4])

#Configura o dataframe resultante
data_trans = pd.DataFrame(data=trans)
data_trans.columns = ['PC1','PC2'] 
data_trans['species'] =  data['species']
```

```python
import matplotlib.pyplot as plt

#Aumentando o tamanho da figura resultante
%matplotlib inline
%config InlineBackend.rc={'figure.figsize': (12, 8)}

# Comparando as classes do conjunto iris usando os componentes principais
plt.plot(data_trans.PC1[0:50], data_trans.PC2[0:50] ,ls='', marker='o')
plt.plot(data_trans.PC1[50:100], data_trans.PC2[50:100] ,ls='', marker='v')
plt.plot(data_trans.PC1[100:150], data_trans.PC2[100:150] ,ls='', marker='s')
```

```python
from sklearn.decomposition import TruncatedSVD

#Cria uma instância da classe SVD
SVD = TruncatedSVD(n_components=2)

#Ajusta a instância aos dados e o transforma
trans = SVD.fit_transform(data.iloc[0:,:4])

#Configura o dataframe resultante
data_trans = pd.DataFrame(data=trans)
data_trans.columns = ['PC1','PC2'] 
data_trans['species'] =  data['species']
```

```python
# Comparando as classes do conjunto iris usando os componentes principais
plt.plot(data_trans.PC1[0:50], data_trans.PC2[0:50] ,ls='', marker='o')
plt.plot(data_trans.PC1[50:100], data_trans.PC2[50:100] ,ls='', marker='v')
plt.plot(data_trans.PC1[100:150], data_trans.PC2[100:150] ,ls='', marker='s')
```

```python
from sklearn.decomposition import KernelPCA
#Cria uma instância da classe KernelPCA
kPCA = KernelPCA(n_components=2, kernel='rbf', gamma=1.0)
#Ajusta a instância aos dados e o transforma
trans = kPCA.fit_transform(data.iloc[0:,:4])
#Configura o dataframe resultante
data_trans = pd.DataFrame(data=trans)
data_trans.columns = ['PC1','PC2'] 
data_trans['species'] =  data['species']
```

```python
plt.plot(data_trans.PC1[0:50], data_trans.PC2[0:50] ,ls='', marker='o')
plt.plot(data_trans.PC1[50:100], data_trans.PC2[50:100] ,ls='', marker='v')
plt.plot(data_trans.PC1[100:150], data_trans.PC2[100:150] ,ls='', marker='s')
```
