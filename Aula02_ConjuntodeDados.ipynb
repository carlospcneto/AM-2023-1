# Aula 2 - Conjunto dos dados
Este notebook contém os exemplos de aula, aplicando sobre o conjunto de dados Iris.

```python
import numpy as np
import pandas as pd

# Localização do arquivo
filepath = 'Iris_Data.csv'
# Importando os dados
data = pd.read_csv(filepath)
# Imprimindo as cinco primeras linhas
print(data.iloc[:5])
```

```python
# Calculando a área da sépala
data['sepal_area'] = data.sepal_length * data.sepal_width

# Imprimindo algumas linhas e colunas
print(data.iloc[:5, -3:])
```

```python
# Usando o método size com DataFrame 
# Para séries, usar o método .value_counts
group_sizes = (data
              .groupby('species')
              .size())
print(group_sizes)
```

```python
# Média sobre o DataFrame
print(data.mean())
# Média sobre série
print(data.petal_length.median())
# Moda calculada sobre série
print(data.petal_length.mode())
```

```python
# Desvio padrão, variância, e desvio padrão da média 
print(data.petal_length.std(), data.petal_length.var(), data.petal_length.sem())
# Retorna quartis
print(data.quantile(0))
```

```python
#Análise e descrição dos dados
print(data.describe())
```

```python
# Análise da covariância entre os atributos Iris
print(data.cov())
```

```python
#Análise da correlação entre os atributos
print(data.corr())
```

```python
#Fazendo histograma 3D na marra
#Precisa de ajustes e melhoramentos, se alguém puder fazê-lo me avise

petal = np.zeros(9,dtype=int)
length_range = max(data.petal_length) - min(data.petal_length)
width_range = max(data.petal_width) - min(data.petal_width)

for i in range(data.shape[0]):
    if data.petal_length[i] < min(data.petal_length) + length_range*1/3 :
        if data.petal_width[i] < min(data.petal_width) + width_range*1/3:
            petal[0] += 1
        elif data.petal_width[i] < min(data.petal_width) + width_range*2/3:
            petal[1] += 1
        elif data.petal_width[i] < min(data.petal_width) + width_range:
            petal[2] += 1
    elif data.petal_length[i] < min(data.petal_length) + length_range*2/3 :
            if data.petal_width[i] < min(data.petal_width) + width_range*1/3:
                petal[3] += 1
            elif data.petal_width[i] < min(data.petal_width) + width_range*2/3:
                petal[4] += 1
            elif data.petal_width[i] < min(data.petal_width) + width_range:
                petal[5] += 1
    elif data.petal_length[i] < min(data.petal_length) + length_range :
            if data.petal_width[i] < min(data.petal_width) + width_range*1/3:
                petal[6] += 1
            elif data.petal_width[i] < min(data.petal_width) + width_range*2/3:
                petal[7] += 1
            elif data.petal_width[i] < min(data.petal_width) + width_range:
                petal[8] += 1
  
fig = plt.figure()
ax1 = fig.add_subplot(111, projection='3d')
xpos = np.zeros(9)
ypos = np.zeros(9)

xpos[0] = min(data.petal_length)
xpos[3] = min(data.petal_length)
xpos[6] = min(data.petal_length)
xpos[1] = min(data.petal_length) + length_range*1/3
xpos[4] = min(data.petal_length) + length_range*1/3
xpos[7] = min(data.petal_length) + length_range*1/3
xpos[2] = min(data.petal_length) + length_range*2/3
xpos[5] = min(data.petal_length) + length_range*2/3
xpos[8] = min(data.petal_length) + length_range*2/3
ypos[0] = min(data.petal_width)
ypos[3] = min(data.petal_width)
ypos[6] = min(data.petal_width)
ypos[1] = min(data.petal_width) + width_range*1/3
ypos[4] = min(data.petal_width) + width_range*1/3
ypos[7] = min(data.petal_width) + width_range*1/3
ypos[2] = min(data.petal_width) + width_range*2/3
ypos[5] = min(data.petal_width) + width_range*2/3
ypos[8] = min(data.petal_width) + width_range*2/3
zpos = [0,0,0,0,0,0,0,0,0]
dx = [length_range*1/3]*9
dy = [width_range*1/3]*9
dz = petal

ax1.bar3d(xpos, ypos, zpos, dx, dy, dz, color='#00ceaa')
plt.show()    
```

```python
# Categorias e número de variáveis
categories=list(data)[0:4]
N = len(categories)
 
# Monta o ângulo de cada exigo no gráfico 
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]
 
# Inicializa gráfico
ax = plt.subplot(111, polar=True)
 
# Primeiro eixo no topo para visualização
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
 
# Desenha um eixo por variável e rotula
plt.xticks(angles[:-1], categories)
 
# Desenha rótulos de no eixo Y
ax.set_rlabel_position(0)
plt.yticks([2,4,6], ["2","4","6"], color="grey", size=7)
plt.ylim(0,8)
  
# Plotando gráfico 1, exemplo número 5
values=data.loc[5].drop('species').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label="Setosa")
ax.fill(angles, values, 'b', alpha=0.1)
 
# Plotando gráfico 2, exemplo número 105
values=data.loc[105].drop('species').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label="Virginica")
ax.fill(angles, values, 'r', alpha=0.1)
 
# Coloca legenda
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
```
