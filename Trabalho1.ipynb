## Estudo do Conjunto de Dados ##

### Cojunto de dados de diagnóstico de câncer de mama em Wisconsin ###
Classificação da Biópsia de Massa Mamária: Tumores Malignos vs Tumores Benignos

Recursos de biópsia para classificação de 569 massas mamárias malignas (câncer) e benignas (não câncer).

As características foram extraídas computacionalmente de imagens digitais de lâminas de biópsia por aspiração com agulha fina. As características correspondem às propriedades dos núcleos celulares, como tamanho, forma e regularidade. A média, o erro padrão e o pior valor de cada um dos 10 parâmetros nucleares são relatados para um total de 30 atributos.

Disponivel em https://www.kaggle.com/datasets/utkarshx27/breast-cancer-wisconsin-diagnostic-dataset

### Detalhes dos atributos ###

- **y** - Os resultados. Um fator com dois níveis que indica se uma massa é maligna ("M") ou benigna ("B").

- **x** - Os preditores. Uma matriz com a média, erro padrão e pior valor de cada uma das 10 medições nucleares em cada lâmina, totalizando 30 características por biópsia:

* **radius** -  Raio do núcleo (média das distâncias do centro aos pontos na periferia).

- **texture** - Textura do núcleo (desvio padrão dos valores em escala de cinza).

- **perimeter** - Perímetro do núcleo.

- **area** - Área do núcleo.

- **smoothness** - Suavidade do núcleo (variação local nos comprimentos dos raios).

- **compactness** - Compacidade do núcleo (perímetro^2/área - 1).

- **concavity** - Concavidade do núcleo (gravidade das porções côncavas do contorno).

- **concave_pts** - Número de porções côncavas do contorno do núcleo.

- **symmetry** - Simetria do núcleo.

- **fractal_dim** - Dimensão fractal do núcleo (aproximação da "linha da costa" - 1).

- **_mean** - média.

- **_se** - Erro padrão.

- **_worst** - Pior/maior valor.

### Pre-Processamento ###

```python
# Separando entre atributos e classe
X = df.iloc[:,:-1]
y = df.iloc[:, -1]

# Dividindo entre treino e teste (holdout)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
# Tratamento dos dados
# Normalizacao dos dados
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Padronizacao
# Transforma os dados de forma que a média seja 0 e o desvio padrão seja 1.
# Preserva a forma da distribuição original dos dados.
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)

# Normalizacao
# Transforma os dados de forma que eles sejam ajustados em uma faixa específica, geralmente entre 0 e 1.
# Os valores normalizados estarão na escala relativa aos valores mínimos e máximos originais.
# Preserva a ordem relativa dos dados.
minmax = MinMaxScaler()
X_train_minmax = minmax.fit_transform(X_train)
X_test_minmax = minmax.transform(X_test)
```

### Classificacao com KNN ###

```python
# Classificacao com KNN
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

knn_saves = {}
accuracy = 0.0
max_ac = 0.0
for k in range(1,100):
    #Cria uma instância 
    KNN = KNeighborsClassifier(n_neighbors=k)

    #Ajusta a modelo aos dados de treino
    KNN = KNN.fit(X_train_minmax, y_train)    
    y_pred = KNN.predict(X_test_minmax)
    old_ac = accuracy
    accuracy = accuracy_score(y_test, y_pred)*100
    if round(old_ac, 2) != round(accuracy, 2):    
        print(f"KNeighborsClassifier, k = {k}, Acurácia: {accuracy:.3f}%")
    if accuracy > max_ac:
        max_ac = accuracy
        k_max = k
        y_pred_max = y_pred
print(f"KNeighborsClassifier, k = {k_max}, Acurácia maxima foi: {max_ac:.3f}%")
confusion_matrix = confusion_matrix(y_test, y_pred_max)
print(confusion_matrix)
```
