## Estudo do Conjunto de Dados ##

### Cojunto de dados de diagnóstico de câncer de mama em Wisconsin ###
Classificação da Biópsia de Massa Mamária: Tumores Malignos vs Tumores Benignos

Recursos de biópsia para classificação de 569 massas mamárias malignas (câncer) e benignas (não câncer).

As características foram extraídas computacionalmente de imagens digitais de lâminas de biópsia por aspiração com agulha fina. As características correspondem às propriedades dos núcleos celulares, como tamanho, forma e regularidade. A média, o erro padrão e o pior valor de cada um dos 10 parâmetros nucleares são relatados para um total de 30 atributos.

Disponivel em https://www.kaggle.com/datasets/utkarshx27/breast-cancer-wisconsin-diagnostic-dataset

### Detalhes dos atributos ###

- **y** - Os resultados. Um fator com dois níveis que indica se uma massa é maligna ("M") ou benigna ("B").

- **x** - Os preditores. Uma matriz com a média, erro padrão e pior valor de cada uma das 10 medições nucleares em cada lâmina, totalizando 30 características por biópsia:

* **radius** -  Raio do núcleo (média das distâncias do centro aos pontos na periferia).

- **texture** - Textura do núcleo (desvio padrão dos valores em escala de cinza).

- **perimeter** - Perímetro do núcleo.

- **area** - Área do núcleo.

- **smoothness** - Suavidade do núcleo (variação local nos comprimentos dos raios).

- **compactness** - Compacidade do núcleo (perímetro^2/área - 1).

- **concavity** - Concavidade do núcleo (gravidade das porções côncavas do contorno).

- **concave_pts** - Número de porções côncavas do contorno do núcleo.

- **symmetry** - Simetria do núcleo.

- **fractal_dim** - Dimensão fractal do núcleo (aproximação da "linha da costa" - 1).

- **_mean** - média.

- **_se** - Erro padrão.

- **_worst** - Pior/maior valor.

```python
combine_atts = []
for att in attributes_col_mean[4:]:
    arr_att = attributes_col_mean[0:3]
    arr_att.append(att)
    combine_atts.append(arr_att)
col_pca = df_pca.columns.to_list()[:-1]
sub_df = {
    "all" : df[columns[:-1]], # Todos os atributos
    "mean" : df[attributes_col_mean], # Todos os atributos de média
    "worst" : df[attributes_col_worst], # Todos os atritubos de pior caso
    "mean_rpa" : df[['x.radius_mean', 'x.perimeter_mean', 'x.area_mean']], # Os tres atributos escolhidos visualmente média
    "mean_rpaworst" : df[['x.radius_worst', 'x.perimeter_worst', 'x.area_worst']], # Os tres atributos escolhidos visualmente pior caso
    "mean_rpasmoot" : df[combine_atts[0]], # + smoothness_mean
    "mean_rpacompa" : df[combine_atts[1]], # + compactness_mean
    "mean_rpaconcav" : df[combine_atts[2]], # + concavity_mean
    "mean_rpaconcpt" : df[combine_atts[3]], # + concave_pts_mean
    "mean_rpasimme" : df[combine_atts[4]], # + symmetry_mean
    "mean_rpafract" : df[combine_atts[5]], # + fractal_dim_mean
    "pca" : df_pca[col_pca] # PCA
    }

results_geral = []

# and df_pca - Principal Component Analysis
```

### Classificacao com KNN ###

```python
#Impots 
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, roc_auc_score

for sub_key in sub_df:
    print(f"Classificando o conjunto com Split Train KNN: {sub_key}")
    # Separando entre atributos e classe
    X = sub_df[sub_key].iloc[:]
    y = df.iloc[:, -1]

    # Dividindo entre treino e teste
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)

    # Normalizacao dos dados
    minmax = MinMaxScaler()
    X_train_minmax = minmax.fit_transform(X_train)
    X_test_minmax = minmax.transform(X_test)

    # Classificacao com KNN
    knn_split_results = []
    # Vou variar o número do K viznhos
    for k in range(1,100):
        #Cria uma instância 
        KNN = KNeighborsClassifier(n_neighbors=k)

        #Ajusta a modelo aos dados de treino
        KNN = KNN.fit(X_train_minmax, y_train)

        # Predizendo os dados
        y_pred = KNN.predict(X_test_minmax)

        # Prever as probabilidades de classe
        y_scores = KNN.predict_proba(X_test_minmax)[:,1]

        # Metricas
        accuracy = accuracy_score(y_test, y_pred)*100
        precision = precision_score(y_test, y_pred, average='weighted')*100
        roc_auc = roc_auc_score(y_test, y_scores)

        knn_split_results.append((k, accuracy, precision, roc_auc, KNN))
    
    result_ord = sorted(knn_split_results, key=lambda x: x[1])
    accuracy = result_ord[0][1]
    k = result_ord[0][0]
    precision = result_ord[0][2]
    roc_auc = result_ord[0][3]
    print(f"Melhor acurácia: {accuracy:.4f}%, para k = {k}, Precisao: {precision:.4f}%, Area sob a curva: {roc_auc:.2f}")
    results_geral.append({'Objeto' : KNN, 'Train Split ou Cross Validation': "Split Train", 'Subconjunto' : sub_key,\
                'Acurácia' : float(accuracy), 'Precisão' : float(precision),\
                'Área sob a curva' : float(roc_auc), "Valor de K" : int(k)})


for sub_key in ["mean", "worst", "mean_rpa", "mean_rpaworst",\
                 "mean_rpasmoot", "mean_rpacompa", "mean_rpaconcav",\
                 "mean_rpaconcpt", "mean_rpasimme", "mean_rpafract", "pca"]:
    print(f"Classificando o conjunto com Cross Validation KNN: {sub_key}")
    # Separando entre atributos e classe
    X = sub_df[sub_key].iloc[:]
    y = df_pca.iloc[:, -1]

    knn_cross_results = []
    for k in range(1, 100):
        # Variando numero de folds
        for r in range(2,10):
            # Cria uma instancia
            KNN = KNeighborsClassifier(n_neighbors=k)

            # Metricas
            accuracy = cross_val_score(KNN, X, y, scoring="accuracy", cv=r).mean()*100
            precision = cross_val_score(KNN, X, y, scoring="precision", cv=r).mean()*100
            roc_auc = cross_val_score(KNN, X, y, scoring="roc_auc", cv=r).mean()

            knn_cross_results.append((k, r, accuracy, precision, roc_auc))

    result_ord = sorted(knn_cross_results, key=lambda x: x[2])
    
    k = result_ord[0][0]
    accuracy = result_ord[0][2]
    r = result_ord[0][1]
    precision = result_ord[0][3]
    roc_auc = result_ord[0][4]

    print(f"Melhor acurácia: {accuracy:.4f}%, para k = {k} e r = {r} , Precisao: {precision:.4f}, Area sob a curva: {roc_auc:.2f}")
    results_geral.append({'Objeto' : KNN, 'Train Split ou Cross Validation': "Cross Validation", 'Subconjunto' : sub_key, \
            'Acurácia' : float(accuracy), 'Precisão' : float(precision),\
            'Área sob a curva' : float(roc_auc), "Valor de K" : int(k),\
            'Quantidade de Folders' : int(r)})
```
