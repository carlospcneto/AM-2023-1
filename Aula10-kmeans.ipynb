```python
import pandas as pd
#Importa kmeans do Sklearn
from sklearn.cluster import KMeans
#importa pyplot
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [10, 5]
# Localização do arquivo
filepath = 'data/Iris_Data.csv'
# Importando os dados
data = pd.read_csv(filepath)

#Cria uma instância de classe kmeans com inicialização kmeans++
kmeans = KMeans(n_clusters=3, init='k-means++')

#aplica k-means sobre os dados, desconsiderando classes
kmeans = kmeans.fit(data.iloc[:,:4])

#imprime rótulos do k-means sobre os dados
print(kmeans.labels_)

# Rótulo dos grupos usando comprimento e largura da pétala
fig = plt.figure()
plt.scatter(data.petal_length, data.petal_width, c=kmeans.labels_)
plt.xlabel('Comprimento Pétala')
plt.ylabel('Largura Pétala')
```

```python
#Colocando os dados em ordem aleatória 
randomdata = (data.sample(n=150, replace=False))
#Aplicando hold out
traindata = randomdata.iloc[:135,:]
testdata = randomdata.iloc[135:,:]

#aplica k-means sobre os dados, desconsiderando classes
kmeans = kmeans.fit(traindata.iloc[:,0:4])

#Ajusta o modelo SVM aos dados de treino
predito = kmeans.predict(testdata.iloc[:,0:4])

print(predito)
```
